This is important to optimize hyperparamter tuning 

```python
#importing dependencies 
import numpy as np
import pandas as pd
import sklearn.datasets as sk
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
data = sk.load_breast_cancer()
```

```python
print(data)
```

```python
data1 = pd.DataFrame(data.data,columns=data.feature_names)
```

```python
data1.head()
```

```python
data1['label'] = data.target
```

```python
data1.head(10)
```

```python
data1.shape
```

```python
data1.isnull().sum()
```

```python
data1.info()
```

```python
data1.describe()
```

```python
data1['label'].value_counts()
```

```python
X = data1.drop('label',axis=1)
Y= data1['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

HyperParameter Tuning 

BY GridSearchCv : -

```python
model = SVC()
```

```python
parameter = {
    'kernel' : ['linear' ,'poly','rbf','sigmoid'],
    'C' : [1,5,10,15,20]
}
```

```python
classifier = GridSearchCV(model,parameter,cv=5)
```

```python
classifier.fit(X,Y)
```

```python
print(classifier.cv_results_)
```

```python
print(classifier.best_params_)
```

```python
print(classifier.best_score_)
```

```python
data2 = pd.DataFrame(classifier.cv_results_)
```

```python
data2.head()
```

```python
result_data = data2[['param_C','param_kernel','mean_test_score']]
```

```python
print(result_data)
```

Using RandomizedSearchCV

```python
model1 = SVC()
```

```python
classifier1 = RandomizedSearchCV(model1,parameter,cv=6)
```

```python
classifier1.fit(X,Y)
```

```python
print(classifier1.cv_results_)
```

```python
print(classifier1.best_params_)
```

```python
print(classifier1.best_score_)
```

```python
data11 = pd.DataFrame(classifier1.cv_results_)
```
